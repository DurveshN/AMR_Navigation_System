**SVG → Bounding Boxes → Visualize**

```python
# ================================
# SVG → Bounding Boxes → Visualize
# ================================

import xml.etree.ElementTree as ET
from svgpathtools import parse_path
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# -------------------------------
# CONFIG
# -------------------------------
SVG_FILE = "map.svg"   # <-- your SVG file
MAP_W = 1000           # canvas width
MAP_H = 1000           # canvas height
MIN_SIZE = 20          # filter small noise

# -------------------------------
# STEP 1: LOAD SVG
# -------------------------------
tree = ET.parse(SVG_FILE)
root = tree.getroot()

raw_objects = []

# -------------------------------
# STEP 2: EXTRACT PATH BOUNDING BOXES
# -------------------------------
for elem in root.iter():
    tag = elem.tag.split('}')[-1]

    if tag == "path":
        d = elem.attrib.get('d')
        if not d:
            continue

        try:
            path = parse_path(d)
            xmin, xmax, ymin, ymax = path.bbox()

            x = xmin
            y = ymin
            w = xmax - xmin
            h = ymax - ymin

            raw_objects.append((x, y, w, h))

        except Exception:
            continue

print(f"Total raw objects: {len(raw_objects)}")

# -------------------------------
# STEP 3: FIX NEGATIVE + CLIP
# -------------------------------
fixed_objects = []

for (x, y, w, h) in raw_objects:

    if w <= 0 or h <= 0:
        continue

    # Convert to corners
    x1 = x
    y1 = y
    x2 = x + w
    y2 = y + h

    # Clip to map boundaries (IMPORTANT)
    x1 = max(0, min(x1, MAP_W))
    y1 = max(0, min(y1, MAP_H))
    x2 = max(0, min(x2, MAP_W))
    y2 = max(0, min(y2, MAP_H))

    # Recalculate width/height
    w = x2 - x1
    h = y2 - y1

    # Remove tiny noise
    if w < MIN_SIZE or h < MIN_SIZE:
        continue

    fixed_objects.append((x1, y1, w, h))

print(f"Filtered objects: {len(fixed_objects)}")

# -------------------------------
# STEP 4: PRINT SAMPLE OUTPUT
# -------------------------------
print("\nSample objects:")
for obj in fixed_objects[:10]:
    print(obj)

# -------------------------------
# STEP 5: VISUALIZATION
# -------------------------------
fig, ax = plt.subplots(figsize=(6, 6))

# Set coordinate system (match SVG)
ax.set_xlim(0, MAP_W)
ax.set_ylim(MAP_H, 0)  # invert Y-axis

# Draw bounding boxes
for (x, y, w, h) in fixed_objects:
    rect = patches.Rectangle(
        (x, y), w, h,
        linewidth=1,
        edgecolor='red',
        facecolor='none'
    )
    ax.add_patch(rect)

# Draw center points
for (x, y, w, h) in fixed_objects:
    cx = x + w / 2
    cy = y + h / 2
    ax.plot(cx, cy, 'bo', markersize=2)

plt.title("Extracted Bounding Boxes from SVG")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(False)

plt.show()
```

**SVG → Clean Objects → Deduplicated Boxes → Visualize -> Occupancy Grid -> INFLATED GRID**

```python
# ==========================================
# SVG → Clean Objects → Deduplicated Boxes → Visualize
# ==========================================

import xml.etree.ElementTree as ET
from svgpathtools import parse_path
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# -------------------------------
# CONFIG
# -------------------------------
SVG_FILE = "map.svg"
MAP_W = 1000
MAP_H = 1000
MIN_SIZE = 20

# Keep only real object colors (VERY IMPORTANT)
VALID_COLORS = {"#9B7C7D", "#D6A45A", "#696969"}

# -------------------------------
# STEP 1: LOAD SVG
# -------------------------------
tree = ET.parse(SVG_FILE)
root = tree.getroot()

raw_objects = []

# -------------------------------
# STEP 2: EXTRACT PATHS (WITH COLOR FILTER)
# -------------------------------
for elem in root.iter():
    tag = elem.tag.split('}')[-1]

    if tag == "path":
        d = elem.attrib.get('d')
        fill = elem.attrib.get('fill')

        # 🔥 Skip noise (most important fix)
        if fill not in VALID_COLORS:
            continue

        if not d:
            continue

        try:
            path = parse_path(d)
            xmin, xmax, ymin, ymax = path.bbox()

            x = xmin
            y = ymin
            w = xmax - xmin
            h = ymax - ymin

            raw_objects.append((x, y, w, h))

        except Exception:
            continue

print(f"Total raw objects (after color filter): {len(raw_objects)}")

# -------------------------------
# STEP 3: CLIP + SIZE FILTER
# -------------------------------
filtered_objects = []

for (x, y, w, h) in raw_objects:

    if w <= 0 or h <= 0:
        continue

    # Convert to corners
    x1 = x
    y1 = y
    x2 = x + w
    y2 = y + h

    # Clip to bounds
    x1 = max(0, min(x1, MAP_W))
    y1 = max(0, min(y1, MAP_H))
    x2 = max(0, min(x2, MAP_W))
    y2 = max(0, min(y2, MAP_H))

    w = x2 - x1
    h = y2 - y1

    # Remove tiny objects
    if w < MIN_SIZE or h < MIN_SIZE:
        continue

    # Remove thin shapes (edges/borders)
    ratio = max(w/h, h/w)
    if ratio > 10:
        continue

    filtered_objects.append((x1, y1, w, h))

print(f"After size + shape filtering: {len(filtered_objects)}")

# -------------------------------
# STEP 4: REMOVE DUPLICATES (IoU)
# -------------------------------
def iou(b1, b2):
    x1, y1, w1, h1 = b1
    x2, y2, w2, h2 = b2

    xa = max(x1, x2)
    ya = max(y1, y2)
    xb = min(x1 + w1, x2 + w2)
    yb = min(y1 + h1, y2 + h2)

    inter = max(0, xb - xa) * max(0, yb - ya)
    union = w1*h1 + w2*h2 - inter

    return inter / union if union > 0 else 0


final_objects = []

for box in filtered_objects:
    keep = True
    for f in final_objects:
        if iou(box, f) > 0.85:
            keep = False
            break

    if keep:
        final_objects.append(box)

print(f"Final unique objects: {len(final_objects)}")

# -------------------------------
# STEP 5: PRINT SAMPLE
# -------------------------------
print("\nFinal objects:")
for obj in final_objects:
    print(obj)

# -------------------------------
# STEP 6: VISUALIZATION
# -------------------------------
fig, ax = plt.subplots(figsize=(6, 6))

ax.set_xlim(0, MAP_W)
ax.set_ylim(MAP_H, 0)

# Draw clean boxes
for (x, y, w, h) in final_objects:
    rect = patches.Rectangle(
        (x, y), w, h,
        linewidth=2,
        edgecolor='red',
        facecolor='none'
    )
    ax.add_patch(rect)

# Draw centers
for (x, y, w, h) in final_objects:
    cx = x + w/2
    cy = y + h/2
    ax.plot(cx, cy, 'bo', markersize=3)

plt.title("CLEANED Bounding Boxes (After Filtering + Dedup)")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

```python
# ==========================================
# STEP 7: BUILD OCCUPANCY GRID
# ==========================================

GRID_SIZE = 100
CELL_SIZE = MAP_W // GRID_SIZE   # ~10 px per cell

# Initialize grid
grid = [[0 for _ in range(GRID_SIZE)] for _ in range(GRID_SIZE)]

# Fill obstacles
for (x, y, w, h) in final_objects:

    gx1 = int(x // CELL_SIZE)
    gy1 = int(y // CELL_SIZE)

    gx2 = int((x + w) // CELL_SIZE)
    gy2 = int((y + h) // CELL_SIZE)

    # Fill cells
    for gy in range(gy1, gy2 + 1):
        for gx in range(gx1, gx2 + 1):

            if 0 <= gx < GRID_SIZE and 0 <= gy < GRID_SIZE:
                grid[gy][gx] = 1
```

```python
# ==========================================
# STEP 8: VISUALIZE GRID
# ==========================================

import numpy as np

grid_np = np.array(grid)

plt.figure(figsize=(6,6))
plt.imshow(grid_np, cmap='gray_r')  # black = obstacle
plt.title("Occupancy Grid (100x100)")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

```python
# ==========================================
#  SAVE OCCUPANCY GRID
# ==========================================

import json

occupancy_data = {
    "width": GRID_SIZE,
    "height": GRID_SIZE,
    "cell_size_px": CELL_SIZE,
    "data": grid
}

with open("occupancy_grid.json", "w") as f:
    json.dump(occupancy_data, f, indent=2)

print("Saved: occupancy_grid.json")
```

```python
# ==========================================
# STEP 9: INFLATION (ROBOT SAFETY)
# ==========================================

INFLATION_RADIUS = 1   # cells (tune this)

inflated = [row[:] for row in grid]

for y in range(GRID_SIZE):
    for x in range(GRID_SIZE):

        if grid[y][x] == 1:

            for dy in range(-INFLATION_RADIUS, INFLATION_RADIUS+1):
                for dx in range(-INFLATION_RADIUS, INFLATION_RADIUS+1):

                    ny = y + dy
                    nx = x + dx

                    if 0 <= nx < GRID_SIZE and 0 <= ny < GRID_SIZE:
                        inflated[ny][nx] = 1
```

```python
plt.figure(figsize=(6,6))
plt.imshow(np.array(inflated), cmap='gray_r')
plt.title("Inflated Occupancy Grid (Safe for Robot)")
plt.show()
```

```python
# ==========================================
#  SAVE INFLATED GRID
# ==========================================

inflated_data = {
    "width": GRID_SIZE,
    "height": GRID_SIZE,
    "cell_size_px": CELL_SIZE,
    "inflation_radius": INFLATION_RADIUS,
    "data": inflated
}

with open("inflated_grid.json", "w") as f:
    json.dump(inflated_data, f, indent=2)

print("Saved: inflated_grid.json")
```

```python
combined = {
    "grid": grid,
    "inflated": inflated,
    "meta": {
        "width": GRID_SIZE,
        "height": GRID_SIZE,
        "cell_size_px": CELL_SIZE,
        "inflation_radius": INFLATION_RADIUS
    }
}

with open("map_data.json", "w") as f:
    json.dump(combined, f, indent=2)
```

```python
# ==========================================
# SVG → Clean Objects → Deduplicated Boxes → Visualize
# ==========================================

import xml.etree.ElementTree as ET
from svgpathtools import parse_path
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# -------------------------------
# CONFIG
# -------------------------------
SVG_FILE = "map.svg"
MAP_W = 1000
MAP_H = 1000
MIN_SIZE = 20

# Keep only real object colors (VERY IMPORTANT)
VALID_COLORS = {"#9B7C7D", "#D6A45A", "#696969"}

# -------------------------------
# STEP 1: LOAD SVG
# -------------------------------
tree = ET.parse(SVG_FILE)
root = tree.getroot()

raw_objects = []

# -------------------------------
# STEP 2: EXTRACT PATHS (WITH COLOR FILTER)
# -------------------------------
for elem in root.iter():
    tag = elem.tag.split('}')[-1]

    if tag == "path":
        d = elem.attrib.get('d')
        fill = elem.attrib.get('fill')

        # 🔥 Skip noise (most important fix)
        if fill not in VALID_COLORS:
            continue

        if not d:
            continue

        try:
            path = parse_path(d)
            xmin, xmax, ymin, ymax = path.bbox()

            x = xmin
            y = ymin
            w = xmax - xmin
            h = ymax - ymin

            raw_objects.append((x, y, w, h))

        except Exception:
            continue

print(f"Total raw objects (after color filter): {len(raw_objects)}")

# -------------------------------
# STEP 3: CLIP + SIZE FILTER
# -------------------------------
filtered_objects = []

for (x, y, w, h) in raw_objects:

    if w <= 0 or h <= 0:
        continue

    # Convert to corners
    x1 = x
    y1 = y
    x2 = x + w
    y2 = y + h

    # Clip to bounds
    x1 = max(0, min(x1, MAP_W))
    y1 = max(0, min(y1, MAP_H))
    x2 = max(0, min(x2, MAP_W))
    y2 = max(0, min(y2, MAP_H))

    w = x2 - x1
    h = y2 - y1

    # Remove tiny objects
    if w < MIN_SIZE or h < MIN_SIZE:
        continue

    # Remove thin shapes (edges/borders)
    ratio = max(w/h, h/w)
    if ratio > 10:
        continue

    filtered_objects.append((x1, y1, w, h))

print(f"After size + shape filtering: {len(filtered_objects)}")

# -------------------------------
# STEP 4: REMOVE DUPLICATES (IoU)
# -------------------------------
def iou(b1, b2):
    x1, y1, w1, h1 = b1
    x2, y2, w2, h2 = b2

    xa = max(x1, x2)
    ya = max(y1, y2)
    xb = min(x1 + w1, x2 + w2)
    yb = min(y1 + h1, y2 + h2)

    inter = max(0, xb - xa) * max(0, yb - ya)
    union = w1*h1 + w2*h2 - inter

    return inter / union if union > 0 else 0


final_objects = []

for box in filtered_objects:
    keep = True
    for f in final_objects:
        if iou(box, f) > 0.85:
            keep = False
            break

    if keep:
        final_objects.append(box)

print(f"Final unique objects: {len(final_objects)}")

# -------------------------------
# STEP 5: PRINT SAMPLE
# -------------------------------
print("\nFinal objects:")
for obj in final_objects:
    print(obj)

# -------------------------------
# STEP 6: VISUALIZATION
# -------------------------------
fig, ax = plt.subplots(figsize=(6, 6))

ax.set_xlim(0, MAP_W)
ax.set_ylim(MAP_H, 0)

# Draw clean boxes
for (x, y, w, h) in final_objects:
    rect = patches.Rectangle(
        (x, y), w, h,
        linewidth=2,
        edgecolor='red',
        facecolor='none'
    )
    ax.add_patch(rect)

# Draw centers
for (x, y, w, h) in final_objects:
    cx = x + w/2
    cy = y + h/2
    ax.plot(cx, cy, 'bo', markersize=3)

plt.title("CLEANED Bounding Boxes (After Filtering + Dedup)")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

```python
# ==========================================
# STEP 6: VISUALIZE WITH UNIQUE COLORS + IDs
# ==========================================

import random

fig, ax = plt.subplots(figsize=(6, 6))

ax.set_xlim(0, MAP_W)
ax.set_ylim(MAP_H, 0)

colored_objects = []

for i, (x, y, w, h) in enumerate(final_objects):

    COLORS = ['red','green','blue','yellow','purple','cyan','orange']
    color = COLORS[i % len(COLORS)]

    # Draw rectangle
    rect = patches.Rectangle(
        (x, y), w, h,
        linewidth=2,
        edgecolor=color,
        facecolor='none'
    )
    ax.add_patch(rect)

    # Label with ID
    ax.text(
        x, y - 5,
        f"id:{i}",
        color=color,
        fontsize=8,
        weight='bold'
    )

    # Draw center
    cx = x + w/2
    cy = y + h/2
    ax.plot(cx, cy, 'o', color=color, markersize=4)

    # Store for later
    colored_objects.append({
        "id": f"obj_{i}",
        "color": color,
        "bbox_px": (x, y, w, h)
    })

plt.title("Objects with Unique IDs (Label these)")
plt.show()
```

```python
# ==========================================
# STEP: BUILD SEMANTIC LAYER (MANUAL LABEL)
# ==========================================

GRID_SIZE = 100
CELL_SIZE = MAP_W // GRID_SIZE

# Your mapping
LABEL_MAP = {
    "drover": [1, 3, 5, 7, 13, 14],
    "bed": [2, 10, 11, 12],
    "table": [6, 8, 9],
    "chair": [4]
}

# Reverse map: id → type
id_to_type = {}
for obj_type, ids in LABEL_MAP.items():
    for i in ids:
        id_to_type[i] = obj_type


semantic_objects = []

for i, (x, y, w, h) in enumerate(final_objects):

    # Convert to grid (IMPORTANT: SAME as occupancy grid)
    gx = int(x // CELL_SIZE)
    gy = int(y // CELL_SIZE)
    gw = max(1, int(w // CELL_SIZE))
    gh = max(1, int(h // CELL_SIZE))

    obj_type = id_to_type.get(i, "unknown")

    semantic_objects.append({
        "id": f"obj_{i}",
        "type": obj_type,
        "bbox": {
            "x": gx,
            "y": gy,
            "w": gw,
            "h": gh
        }
    })


semantic_layer = {
    "objects": semantic_objects
}
```
